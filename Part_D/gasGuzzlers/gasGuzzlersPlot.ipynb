```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/AverageGas.txt', sep = '\t')
df1
```

```python
df1.columns = ['GasPrice']
df1
```

```python
df1[['Date', 'Gasprice']] = df1['GasPrice'].str.split(' ', 1, expand=True)
df1
```

```python
df1['Gasprice'] = df1['Gasprice'].str.strip(' ]')
df1['Date'] = df1['Date'].str.strip('()')
df1['Date'] = df1['Date'].str.replace('"','')
df1['Gasprice'] = df1['Gasprice'].str.strip(')')
df1['Gasprice'] = df1['Gasprice'].astype(float)
#df1['Date'] = df1['Date'].str.replace('.',',')
df1.head()
```

```python
plt.figure(figsize = (30,10))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Prices of Each day')
plt.bar(df1['Date'],  df1['Gasprice'])
```

```python
df1 = pd.read_csv('/content/Timeout.txt', sep = '\t')
df1.head()
```

```python
df1.columns = ['difficulty']
```

```python
df1[['Date','Difficulty & Gasprice']] = df1['difficulty'].str.split(' ', 1, expand=True)
df1
```

```python
df1[['Difficulty','Gasprice']] = df1['Difficulty & Gasprice'].str.split(',', 1, expand=True)
df1
```

```python
df1['Date'] = df1['Date'].str.strip('(')
df1['Gasprice'] = df1['Gasprice'].str.strip(')')
df1['Difficulty']=df1['Difficulty'].str.replace('(','')
df1['Difficulty']=df1['Difficulty'].str.replace(',','')
df1['Gasprice'] = df1['Gasprice'].str.strip(')')
df1['Date']=df1['Date'].str.replace(',','')
df1['Difficulty'] = df1['Difficulty'].astype(float)
df1
```

```python
plt.figure(figsize=(30,10))
plt.title('difficulty vs Time')
plt.xlabel('difficulty')
plt.ylabel('Date')
plt.yticks(rotation=90)
plt.xticks(rotation=90)
plt.plot(df1['Date'], df1['difficulty'])
plt.show()
```
