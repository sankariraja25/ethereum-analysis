```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/PartDfork_2.txt', sep = '\t')
df1
```

```python
df1.columns = ['Date', 'Number of Transactions']
df1
```

```python
df1[['Value', 'Gasprice']] = df1['Number of Transactions'].str.split(' ', 1, expand=True)
df1
```

```python
df1['Gasprice'] = df1['Gasprice'].str.strip(' ]')
df1
```

```python
plt.figure(figsize = (30,10))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Prices of Each day')
plt.bar(df1['Date'],  df1['Gasprice'])
```
