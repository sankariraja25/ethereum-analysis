```python
from google.colab import drive 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
def readcsv(path1,path2):
  part0 = pd.read_csv(path1, header=None, sep = '\t')
  part1 = pd.read_csv(path2, header=None, sep = '\t')
  table = part1.append(part0, ignore_index=True)
  return table
```

```python
partA1_table = readcsv('/content/gdrive/MyDrive/etherum/Part_A/PartA1_out/part-00000','/content/gdrive/MyDrive/etherum/Part_A/PartA1_out/part-00001')
partA1_table = partA1_table.sort_values(0)
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('[Month,Year]')
plt.ylabel('No. of Transactions')
plt.bar(partA1_table[0],  partA1_table[1])
```
